Matrix decompositions are most essential in practical cases. There are major benefits of using decomposed/approximated matrix rather than using the actual one

* In most of the cases, the matrices are sparse. Hence, decomposed dense matrices can be used to reduce the runtime
* In many applications like recommender system, the actual matrix size can be very huge to be fitted in the main memory. Decomposed smaller matrices can be used to reduce the storage
* In image analysis, raw image can contain cells which are noisy and contain very less information. Random projection or decompositions can be used to reduce the dimension of the image. The lesser the dimension, the easier the task

```python
import pandas as pd
import scipy
import numpy as np
import matplotlib.pyplot as plt
from sklearn.decomposition import *
from sklearn.random_projection import *
```

We will start with two matrices and check the eligibility and effect of the popular decompositions

```python
M = np.random.randint(1000,size=(4,5))
print (M)
```

```python
(Q,R)=np.linalg.qr(M)
print (Q)
print (R)
```

We see that the matrix has more columns than rows which means we have dependence between columns. Hence, the orthogonal basis contains only 4 elements (rank). The 5th column in R matrix can be written in terms of other columns.

```python
(_Q,_R)=np.linalg.qr(np.transpose(M))
print (_Q)
print (_R)
```

QR decomposition is used to orthogonalize the basis. Moreover, the basis elements are not only orthogonal but orthonormal. 

```python
(_,L,U)= scipy.linalg.lu(M)
print (L)
print (U)
```

```python
(_,_L,_U)= scipy.linalg.lu(np.transpose(M))
print (_L)
print (_U)
```

LU decompositions are analogue to gaussian elimination method we use to solve linear system. Converting matrix into lower and upper triangular matrices are helpful as the inverse of triangular matrices are easier than the full matrix.

Now let us construct a symmetric matrix and check decomposition

```python
S = np.matmul(S,np.transpose(S))
print (S)
```

```python
(Lambda,_S) = np.linalg.eig(S)
print (Lambda)
print (_S)
```

```python
np.matmul(_S, np.transpose(_S))
```

```python
np.linalg.det(np.matmul(_S, np.transpose(_S)))
```

The eigen vectors are orthonormal. This is called Spectral theorem.

```python
scipy.linalg.schur(S)
```

```python
scipy.linalg.schur(N)
```

Schur's theorem says that every square $n \times n$ matrix can be triangularized i.e. can be written in form of $QTQ^{-1}$, where $Q$ is orthnormal and $T$ is upper triangular. This means any symmetric matrix can be diagonalized or, we can apply spectal theorem for any symmetric matrix. 

```python
scipy.linalg.cholesky(S)
```

Cholesky decomposition only holds when the eigen values are > 0 for symmetric matrices. This can be derived from pivot elements as well. In order to have all eigen values > 0, all the pivot elements has to be > 0

```python
S[2,2] = 2
scipy.linalg.cholesky(S)
```

```python
S[2,2] = -2
scipy.linalg.cholesky(S)
```

```python
from scipy.ndimage import imread
im = imread('dog.jpg',flatten=True)
print (im.shape)
```

```python
(U,D,V) = scipy.linalg.svd(im)
print (U.shape, D.shape, V.shape)
```

```python
def reshape(mat1,mat2):
    dim = mat2.shape[0]-mat1.shape[1]
    mat = np.zeros((mat1.shape[0],dim))
    mat1 = np.concatenate([mat1,mat],axis=1)
    return mat1
```

```python
_U = U[:,:-100]
_D = D[:-100]
_D = np.diag(_D)
_V = V[:-100,:]
_D = reshape(_D,_V)
im_degen = np.matmul(_U,np.matmul(_D,_V))
print ("Frobenius norm of the error matrix is %.3f" % (np.linalg.norm(im-im_degen,ord=2)))
plt.imshow(im_degen)
plt.show()
```

```python
_U = U[:,:-300]
_D = D[:-300]
_D = np.diag(_D)
_V = V[:-300,:]
_D = reshape(_D,_V)
im_degen = np.matmul(_U,np.matmul(_D,_V))
print ("Frobenius norm of the error matrix is %.3f" % (np.linalg.norm(im-im_degen,ord=2)))
plt.imshow(im_degen)
plt.show()
```
