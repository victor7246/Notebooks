When we start learning mathematics, we first start with numbers aka natural numbers $\mathbb{N}$ (which follows the rule of counting). Then graduatally, we construct rational numbers (fractions). Theories of convergence lead us to irrational numbers and the whole set (field) of real numbers $\mathbb{R}$.
In basic algebra, we learn that a vector space over a field is a set of numbers which follows certain conditions. Thus we construct vectors in higher dimensional space $\mathbb{R}^n$. Each element (vector) of the higher dimensional space looks like $(v_1, v_2, ..., v_n)$.

Next we define linear operators that transform one vector into another. 

$$f:\mathbb{R}^m \rightarrow \mathbb{R}^n$$
Linear operators will form an array of arrays aka matrix, which consist the individual coefficients. Each $n \times m$ matrix look like - 

$$\begin{bmatrix}
    x_{11}       & x_{12} & x_{13} & \dots & x_{1m} \\
    x_{21}       & x_{22} & x_{23} & \dots & x_{2m} \\
    \dots \\
    x_{n1}       & x_{n2} & x_{n3} & \dots & x_{nm}
\end{bmatrix}
$$

We also encounter matrices when we deal with multiple vectors from a vector space. E.g. - an image is nothing but a well aligned matrix of RGB values. The number of elements in vertical strips denotes the dimension of the space.

As we have already discussed about rank of a matrix, let us go through one interesting property of matrices. The four fundamental subspaces of matrix $A$ are $colspace(A)$, $nullspace(A)$, $rowspace(A)$ and $nullspace(A^T)$. Let us see the relationship between these 4 subspaces.

$$A:\mathbb{R}^n \rightarrow \mathbb{R}^m$$

$$\begin{bmatrix}
    x_{11}       & x_{12} & x_{13} & \dots & x_{1n} \\
    x_{21}       & x_{22} & x_{23} & \dots & x_{2n} \\
    \dots \\
    x_{m1}       & x_{m2} & x_{m3} & \dots & x_{mn}
\end{bmatrix}
$$

$Colspace(A)$ is the subspace spanned by the columns of $A$. The columns are in $\mathbb{R}^m$ space. But the dimension of the subspace depends on the number of independent columns of $A$ which is same as $rank(A)$. Similarly, $rowspace(A)$ or $colspace(A^T)$ is a subspace of $\mathbb{R}^n$ of dimension same as $rank(A^T)$. 
On the other hand, $nullspace(A)$ is a subspace of $\mathbb{R}^n$ and $nullspace(A^T)$ is a subspace of $\mathbb{R}^m$.

It can be shown mathematically that $rank(A) = rank(A^T)$. One intuitive proof can be derived from the row echolon form of matrix. In row echolon reduced form, each pivot element has all the below elements as 0. Also, each pivot element has all the left row elements as 0. Hence, number of independent columns is same as number of independent rows. 

Now, if $rank(A) = rank(A^T) = r$, that leads us to $nullity(A) = n-r$ and $nullity(A^T) = m-r$.

Simple mathematics can show that the $colspace(A) \perp nullspace(A^T)$ and $rowspace(A) \perp nullspace(A)$.

We have established that, for any element $x \in rowspace(A)$ and $x_0 \in nullspace(A)$, the matrix sends $x+x_0$ to $Ax$ which is in $colspace(A)$. The below image illustrates the understanding.

In the next section we discuss another popular use case of linear algebra - solving linear equations. Say we have few equations to solve.

$$
x_1 + 2x_2 - 3x_3 = 1 \\
5x_1 - x_2 + x_3 = 9 \\
2x_1 - 6x_2 + 5x_3 = 4
$$

```python
M = np.array([[1,2,-3],[5,-1,1],[2,-6,5]])
print (M)
print (np.linalg.matrix_rank(M))
```

```python
Y = np.array([1,9,4])
print (np.linalg.solve(M,Y))
X = np.matmul(np.linalg.inv(M),Y)
print (X)
```

any matrix $M$ can be decomposed into lower and upper triangular matrix $M = LU$ using gauss elimination. Solving an equation from triangular matrix is much easier. A typical matrix of size $n \times n$ needs $n^3$ time to solve a system of equation. Where as decomposing it into $LU$ takes $\frac{2}{3}(n^3 - n)$ time and $n^2$ time to solve from triangular matrices.

The $3 \times 3$ matrix has 3 independent columns i.e. rank of 3. The matrix is invertible and hence we have a perfect solution. Next we consider a matrix which is wider i.e. number of columns (variables) greater than number of rows (equations). In this situation we have infinite solutions.

$\begin{bmatrix}
    1       & -1 & 3 & 17 \\
    2       & 3 & 11 & -3 \\
   -5       & 0 & 1 & 10 \\
\end{bmatrix}
\begin{bmatrix}
    x_1 \\
    x_2 \\
    x_3 \\
    x_4
\end{bmatrix}
=
\begin{bmatrix}
    1 \\
    3 \\
    10 
\end{bmatrix}
$

```python
M = np.array([[1,-1,3,17],[2,3,11,-3],[-5,0,1,10]])
Y = np.array([1,3,10])
print (M)
print (np.linalg.matrix_rank(M))
```

```python
np.linalg.solve(M,Y)
```

We are unable to solve using elimination method, as, there is no unique solution. Let us find all the solutions.

```python
aug_M = np.append(M,Y.reshape(-1,1),axis=1)
print (aug_M)
```

```python
(P,L,U) = scipy.linalg.lu(aug_M)
```

```python
M_reduced = U[:,:-1]
Y_reduced = U[:,-1]
print (M_reduced, Y_reduced)
```

We see that the last column of M_reduced can be written in terms of other 3 columns. Hence, this column is a free column and other three columns are pivot columns. To get all the possible solutions of the system of equation we need to get the null space and the solution based on the pivot elements. Let us find out the basis for the null space i.e. solution of $M x = 0$

```python
null_x = np.linalg.solve(M_reduced[:,:-1],-M_reduced[:,-1])
print (null_x)
```

```python
x = np.linalg.solve(M_reduced[:,:-1],Y_reduced)
print (x)
```

As the fourth column is free, the null space is of the form $num*[nullx,1]$ for any $num$. The actual solution is the solution based on the first 3 columns $x$. 

Hence, all the possible solutions are $[-1.84761905,-0.56190476 ,0.76190476, 0] + x[1.44761905 ,10.16190476 ,-2.76190476,1]$. Let us see if they are actually solving the equation or not.

```python
for i in range(5):
    num = random.random()
    xdash = np.append(x,0) + num*np.append(null_x,1)
    print (xdash)
    print (np.matmul(M,xdash))
```

tada!! 

In the next example, let us take a matrix which is longer i.e. number of rows (equations) is greater than number of columns (variables). In this situtation, we can have atmost 1 solution. If the rows are consistent, we have an unique solution, else, we do not have any solution.

```python
M = M.transpose()
print (M)
```

```python
Y = np.append(Y,1)
print (Y)
```

```python
aug_M = np.append(M,Y.reshape(-1,1),axis=1)
print (aug_M)
(P,L,U) = scipy.linalg.lu(aug_M)
print (U)
X = np.linalg.solve(U[:-1,:-1],U[:-1,-1])
print (X)
print (np.matmul(M,X))
```

The system of equation is inconsistent due to the last row. As, $0x_1 + 0x_2 + 0x_3 \neq .52296157$ for all $x_i$s

Now let us take a consistent pair of equations.

```python
M[-1,:] = [3,16,-4]
Y = np.array([1,3,10,14])
print (M, Y)
```

```python
aug_M = np.append(M,Y.reshape(-1,1),axis=1)
print (aug_M)
(P,L,U) = scipy.linalg.lu(aug_M)
print (U)
X = np.linalg.solve(U[:-1,:-1],U[:-1,-1])
print (X)
print (np.matmul(M,X))
```

And we are able to generate the same $Y$ from $X$. In this tutorial we have encountered the concept of projection. Projection of a vector $v$ on another vector $u$ is the resultant vector $v^{'}$ which minimized the distance between $\lVert v-u \rVert$. This can be generalized to any two matrices. Next we will see another popular application of projection.

We have seen the difficulty in solving linear equation when the matrix in longer. In case of regression problems, we face the same issue. In any regression problem, we have some independent variables (features), we have the dependent variable (target) and many equations connecting them (correspondence between the features and target). Let us assume that the best fit is a linear model i.e. a straight line. The learning problem is to figure out the coefficients for which

$$
a_1 x_{1} + a_2 x_{2} + \dots + a_n x_{n} = y 
$$

for every x and y in the dataset. The learning objective can be written as $Xa = Y$. We need to figure out optimal $a$.

If $X$ is full-rank invertible matrix, then we can have $a = X^{-1}Y$, which is a closed perfect solution. Also, if $Y$ is in $colspace(X)$ we could find an unique solution. However, the difficulty comes as we often have more equations than the variables and the $Y$ is not in $colspace(X)$. In the previous section, we saw that inconsistency can lead to zero solution. Because we do not have any exact solution, we are interested into finding an approximate solution. In this case we find such $a$ for which $\lVert Xa-Y \rVert$ is minimized. This is also called least square estimate.

The optimal strategy is to find the $a$ for which $Xa$ is the projection of $Y$ on $X$. From the above formula we get - 

$$a = (X^{T}X)^{-1}X^{T}Y$$

Let us try this out with the above example where we had inconsistent equations.

```python
X = np.array([[ 1 ,2,-5],[-1,3 ,0],[ 3,11 ,1], [17, -3, 10]])
Y = np.array([1,3,10,1])
```

After solving the inconsistent pair of equations using $LU$ decomposition we got $a =  [ 0.11246485 ,0.86316776 ,0.16776007]$ and the predicted $Y^{'} = [  1,2.47703843 ,10,1]$ which is different from the actual $Y$. Let us see if the projection gives us the same values.

```python
a = np.matmul(np.matmul(np.linalg.inv(np.matmul(np.transpose(X),X)),np.transpose(X)),Y)
y_predicted = np.matmul(X,a)
print ('esimated a is ', a)
print ('predicted y is ', y_predicted)
```

Okay! So we have got something different. Let us see the errors. We use L2 norm to check the difference predicted y and actual y.

```python
np.linalg.norm(ord=2,x = y_predicted-Y)
```

```python
np.linalg.norm(ord=2,x = np.array([1,2.47703843,10,1])-Y)
```

We see that the least square estimates less prediction error. This illustrates another beautiful and practical application of projections.

In the next section we will discuss about another popular and important application of linear algebra - Eigenvalue decomposition. An eigenvalue $\lambda$ and eigenvector $v$ satisfies $Mv = \lambda v$ .

If all the eigenvalues of a matrix are independent we have a decomposition $M = V\Lambda V^{-1}$

```python
M = np.random.random((2,2))
```

```python
X = []
Y = []
for i in range(5000):
    x = random.random()
    X += [x,x,-x,-x]
    y = np.sqrt(1-x**2)
    Y += [y,-y,y,-y]
    
plt.scatter(X,Y)
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title('Circle')
plt.show()
```

```python
transformed = np.matmul(M,np.vstack([X,Y]))
plt.scatter(transformed[0],transformed[1])
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title('Transformed through M')
plt.show()
```

```python
transformed = np.matmul(M,np.matmul(M,np.vstack([X,Y])))
plt.scatter(transformed[0],transformed[1])
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title('Transformed through M^2')
plt.show()
```

```python
transformed = np.matmul(M,np.matmul(M,np.matmul(M,np.vstack([X,Y]))))
plt.scatter(transformed[0],transformed[1])
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title('Transformed through M^3')
plt.show()
```

More the times we multiply with $M$, the tranformed space becomes to squeezed but the directions of the axes of the ellipse remains same. This is because, $M = V\Lambda V^{-1}$ means $M^2 = V\Lambda^2 V^{-1}$, $M^3 = V\Lambda^3 V^{-1}$

The direction of the transformed space is determined by $V$ but the magnitude is determined by the powers of eigenvalues. As the eigenvalues for this matrix is $<1$, our transformed space become narrower. Let us test our understanding by mulitplying $V$ and $\Lambda$ individually.

```python
(lamdba, v) = np.linalg.eig(M)
print (lamdba)
print (np.linalg.det(v))
```

```python
transformed = np.matmul(np.linalg.inv(v),np.vstack([X,Y]))
plt.scatter(transformed[0],transformed[1])
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title('Transformed through V inverse')
plt.show()
```

```python
transformed = np.matmul(np.diag(lamdba),transformed)
plt.scatter(transformed[0],transformed[1])
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title('Transformed through lambda')
plt.show()
```

```python
transformed = np.matmul(v,transformed)
plt.scatter(transformed[0],transformed[1])
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title('Transformed through V')
plt.show()
```
