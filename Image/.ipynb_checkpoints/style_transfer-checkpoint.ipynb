```python
for i in range(grads_values.shape[0]):
    layer1_grad[:,:,i] = layer1_grad[:,:,i] * grads_values[i]
```

```python
images_per_row=16
nb_cols = grads_values.shape[0] // images_per_row
size = layer1_grad.shape[1]
grid = np.zeros((layer1_grad.shape[1]*nb_cols, layer1_grad.shape[1]*images_per_row))
scale = 1./layer1_grad.shape[1]
def generate_images(array):
    for col in range(nb_cols):
        for row in range(images_per_row):
            feature_map = array[:,:,col*images_per_row + row]
            feature_map -= feature_map.mean()
            feature_map /= feature_map.std()
            feature_map *=255
            feature_map = np.clip(feature_map, 0, 255).astype(np.uint8)
            grid[col*size:(col+1)*size, row*size:(row+1)*size] = feature_map
    
    return grid
```

We could see the more the depth, the more abstract the output. This is because, initial layers learn basic features and later layers learn more abstract feature that is associcated with the classification problem. Hence, on a real dataset these are the tips -

* If the dataset is small and similar to imagenet dataset, use pretrain network removing final layer. Freeze the pretrain model.
* If the dataset is large and similar to imagenet dataset, use pretrain network removing final layer and learn the whole network with small learning rate
* If the dataset is small and different from imagenet dataset, extract network from early layer and start training from new layers
* If the dataset is large and different from imagenet dataset, extract network from early layer and train the whole network

Next we jump onto transfer learning. This notebook is hugely influenced by original style transfer notebook - https://github.com/tensorflow/models/blob/master/research/nst_blogpost/4_Neural_Style_Transfer_with_Eager_Execution.ipynb

Briefly, style transfer depends on two things - 

The loss function between content image and generated image and the loss function between generated image and style image. Two loss functions are added with weights $\alpha$ and $\beta$. 

```python
style_layers = [ 
                'block4_conv1', 
                'block5_conv1'
               ]
```

```python
layer_outputs = [model.get_layer(layer).output for layer in content_layers]
content_model = keras.models.Model(inputs=model.input, outputs=layer_outputs)
```

```python
layer_outputs = [model.get_layer(layer).output for layer in style_layers]
style_model = keras.models.Model(inputs=model.input, outputs=layer_outputs)
```

```python
def content_loss(content_image, generated_image, content_model=content_model):
    generated_image = generated_image.reshape(content_image.shape[0],content_image.shape[1],content_image.shape[2],content_image.shape[3])
    content_outputs = content_model.predict(style_image)[0,:,:,:]
    generated_outputs = content_model.predict(generated_image)[0,:,:,:]
    content_outputs = content_outputs.reshape(content_outputs.shape[0]*content_outputs.shape[1],content_outputs.shape[2])
    generated_outputs = generated_outputs.reshape(generated_outputs.shape[0]*generated_outputs.shape[1],generated_outputs.shape[2])
    return np.mean(np.square(content_outputs - generated_outputs))
```

```python
def grad_content_loss(content_image, generated_image, content_model=content_model):
    generated_image = generated_image.reshape(content_image.shape[0],content_image.shape[1],content_image.shape[2],content_image.shape[3])
    content_outputs = content_model.predict(style_image)[0,:,:,:]
    generated_outputs = content_model.predict(generated_image)[0,:,:,:]
    content_outputs = content_outputs.reshape(content_outputs.shape[0]*content_outputs.shape[1],content_outputs.shape[2])
    generated_outputs = generated_outputs.reshape(generated_outputs.shape[0]*generated_outputs.shape[1],generated_outputs.shape[2])
    return np.mean(content_outputs - generated_outputs)
```

```python
def gram_matrix(matrix):
    return np.matmul(matrix,np.transpose(matrix))
```

```python
def style_loss(style_image, generated_image, style_model=style_model, style_layer_number=len(style_layers)):
    generated_image = generated_image.reshape(style_image.shape[0],style_image.shape[1],style_image.shape[2],style_image.shape[3])
    style_outputs = style_model.predict(style_image)
    generated_outputs = style_model.predict(generated_image)
    loss = 0
    for i in range(style_layer_number):
        style_outputs_layer = style_outputs[i][0,:,:,:]
        generated_outputs_layer = generated_outputs[i][0,:,:,:]
        style_outputs_layer = style_outputs_layer.reshape(style_outputs_layer.shape[0]*style_outputs_layer.shape[1],style_outputs_layer.shape[2])
        generated_outputs_layer = generated_outputs_layer.reshape(generated_outputs_layer.shape[0]*generated_outputs_layer.shape[1],generated_outputs_layer.shape[2])
        loss += np.mean(np.square(gram_matrix(style_outputs_layer)-gram_matrix(generated_outputs_layer)))
        loss /= (2*generated_outputs_layer.shape[0]*generated_outputs_layer.shape[1])**2
    return loss*1.0/style_layer_number
```

```python
def grad_style_loss(style_image, generated_image, style_model=style_model, style_layer_number=len(style_layers)):
    generated_image = generated_image.reshape(style_image.shape[0],style_image.shape[1],style_image.shape[2],style_image.shape[3])
    style_outputs = style_model.predict(style_image)
    generated_outputs = style_model.predict(generated_image)
    loss = 0
    for i in range(style_layer_number):
        style_outputs_layer = style_outputs[i][0,:,:,:]
        generated_outputs_layer = generated_outputs[i][0,:,:,:]
        style_outputs_layer = style_outputs_layer.reshape(style_outputs_layer.shape[0]*style_outputs_layer.shape[1],style_outputs_layer.shape[2])
        generated_outputs_layer = generated_outputs_layer.reshape(generated_outputs_layer.shape[0]*generated_outputs_layer.shape[1],generated_outputs_layer.shape[2])
        loss += np.mean(np.matmul(np.transpose(generated_outputs_layer),gram_matrix(style_outputs_layer)-gram_matrix(generated_outputs_layer)))
        loss /= (2*generated_outputs_layer.shape[0]*generated_outputs_layer.shape[1])**2
    return loss*1.0/style_layer_number
```

```python
alpha = 1e-4
beta = 1e-2
```

```python
def total_loss(generated_image):
    loss1 = content_loss(content_image, generated_image, content_model)
    loss2 = style_loss(style_image, generated_image, style_model, len(style_layers))
    return alpha*loss1 + beta*loss2
```

```python
def grad_total_loss(generated_image):
    loss1 = grad_content_loss(content_image, generated_image, content_model)
    loss2 = grad_style_loss(style_image, generated_image, style_model, len(style_layers))
    return alpha*loss1 + beta*loss2
```

```python
def root_fn(generated_image):
    return [grad_total_loss(generated_image)]*len(generated_image)
```

As it seems, solving such non convex function using scipy is impossible. We will next use tensorflow 

```python
def content_loss(content_image, generated_image, content_model=content_model):
    #content_outputs = content_model.predict(style_image)[0,:,:,:]
    #generated_outputs = content_model.predict(generated_image,steps=1)[0,:,:,:]
    content_outputs  = content_model(tf.convert_to_tensor(content_image))
    generated_outputs  = content_model(generated_image)
    #content_outputs = content_outputs.reshape(content_outputs.shape[0]*content_outputs.shape[1],content_outputs.shape[2])
    #generated_outputs = generated_outputs.reshape(generated_outputs.shape[0]*generated_outputs.shape[1],generated_outputs.shape[2])
    return tf.reduce_mean(tf.square(content_outputs - generated_outputs))
```

```python
def gram_matrix(matrix):
    return tf.matmul(matrix,matrix,transpose_a=True)
```

```python
def style_loss(style_image, generated_image, style_model=style_model, style_layer_number=len(style_layers)):
    #style_outputs = style_model.predict(style_image)
    #generated_outputs = style_model.predict(generated_image,steps=1)
    style_outputs = style_model(tf.convert_to_tensor(style_image))
    generated_outputs = style_model(generated_image)
    global generated_outputs_layer
    loss = 0
    for i in range(style_layer_number):
        style_outputs_layer = style_outputs[i][0,:,:,:]
        generated_outputs_layer = generated_outputs[i][0,:,:,:]
        #style_outputs_layer = style_outputs_layer.reshape(style_outputs_layer.shape[0]*style_outputs_layer.shape[1],style_outputs_layer.shape[2])
        #generated_outputs_layer = generated_outputs_layer.reshape(generated_outputs_layer.shape[0]*generated_outputs_layer.shape[1],generated_outputs_layer.shape[2])
        loss += tf.reduce_mean(tf.square(gram_matrix(style_outputs_layer)-gram_matrix(generated_outputs_layer)))
        height, width, channels = generated_outputs_layer.get_shape().as_list()
        loss /= (4. * (channels ** 2) * (width * height) ** 2)
    return loss*1.0/style_layer_number
```

```python
def total_loss(generated_image):
    loss1 = content_loss(content_image, generated_image, content_model)
    loss2 = style_loss(style_image, generated_image, style_model, len(style_layers))
    return alpha*loss1 + beta*loss2
```

```python
def compute_grads(cfg):
    with tf.GradientTape() as tape: 
        loss = total_loss(cfg['init_image'])
    return tf.gradients(loss, cfg['init_image'])
```

```python
init_image = generated_image
init_image = tf.Variable(init_image,dtype=tf.float32)
```

```python
cfg = {}
cfg['init_image'] = init_image
opt = tf.train.AdamOptimizer(learning_rate=5, beta1=0.99, epsilon=1e-1)
numiters = 500
```

```python
for i in range(numiters):
    grads = compute_grads(cfg)
    print (grads)
    opt.apply_gradients([(grads, init_image)])
```
